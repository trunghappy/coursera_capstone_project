**Define function to covert latitude, longitude to xy coordiante, distance**

```
def lonlat_to_xy(lon, lat):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=33, datum='WGS84')
    xy = pyproj.transform(proj_latlon, proj_xy, lon, lat)
    return xy[0], xy[1]

def xy_to_lonlat(x, y):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=33, datum='WGS84')
    lonlat = pyproj.transform(proj_xy, proj_latlon, x, y)
    return lonlat[0], lonlat[1]

def calc_xy_distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return math.sqrt(dx*dx + dy*dy)

print('Coordinate transformation check')
print('-------------------------------')
print('BinhTan, HCM center longitude={}, latitude={}'.format(hcm_center[1], hcm_center[0]))
x, y = lonlat_to_xy(hcm_center[1], hcm_center[0])
print('BinhTan, HCM  center UTM X={}, Y={}'.format(x, y))
lo, la = xy_to_lonlat(x, y)
print('BinhTan, HCM  center longitude={}, latitude={}'.format(lo, la))
```

```
def compute_average_cummulative(x):
  result = []
  for i in range(len(x)-1):
    average_val = (x[i] + x[i+1])/2
    result.append(average_val)
  return result
```

```
def make_grid_circle_lower(x_range, y_range, distance,num_rows):
  final_x_lower = [x_range]
  final_y_lower = [y_range]
  step = distance * math.sqrt(3) / 2
  stop_point = 0
  while stop_point <= num_rows: 
    x_range_new = [i + step for i in final_x_lower[stop_point][1:]]
    y_range_new = compute_average_cummulative(final_y_lower[stop_point])
    final_x_lower.append(x_range_new)
    final_y_lower.append(y_range_new)
    stop_point +=1
  return final_x_lower[1:], final_y_lower[1:]
```

```
# Find upper circles
def make_grid_circle_upper(x_range, y_range, distance,num_rows):
  final_x_upper = [x_range]
  final_y_upper = [y_range]
  step = distance * math.sqrt(3) / 2
  stop_point = 0
  while stop_point <= num_rows: 
    x_range_new = [i - step for i in final_x_upper[stop_point][1:]]
    y_range_new = compute_average_cummulative(final_y_upper[stop_point])
    final_x_upper.append(x_range_new)
    final_y_upper.append(y_range_new)
    stop_point +=1
  return final_x_upper[1:], final_y_upper[1:]
```
