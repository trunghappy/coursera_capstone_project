**Define function to covert latitude, longitude to xy coordiante, distance**

```
def lonlat_to_xy(lon, lat):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=33, datum='WGS84')
    xy = pyproj.transform(proj_latlon, proj_xy, lon, lat)
    return xy[0], xy[1]

def xy_to_lonlat(x, y):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=33, datum='WGS84')
    lonlat = pyproj.transform(proj_xy, proj_latlon, x, y)
    return lonlat[0], lonlat[1]

def calc_xy_distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return math.sqrt(dx*dx + dy*dy)

print('Coordinate transformation check')
print('-------------------------------')
print('quan 10, HCM center longitude={}, latitude={}'.format(hcm_center[1], hcm_center[0]))
x, y = lonlat_to_xy(hcm_center[1], hcm_center[0])
print('quan 10, HCM  center UTM X={}, Y={}'.format(x, y))
lo, la = xy_to_lonlat(x, y)
print('quan 10, HCM  center longitude={}, latitude={}'.format(lo, la))
```

```
def compute_average_cummulative(x):
  result = []
  for i in range(len(x)-1):
    average_val = (x[i] + x[i+1])/2
    result.append(average_val)
  return result
```

```
def make_grid_circle_lower(x_range, y_range, distance,num_rows):
  final_x_lower = [x_range]
  final_y_lower = [y_range]
  step = distance * math.sqrt(3) / 2
  stop_point = 0
  while stop_point <= num_rows: 
    x_range_new = [i + step for i in final_x_lower[stop_point][1:]]
    y_range_new = compute_average_cummulative(final_y_lower[stop_point])
    final_x_lower.append(x_range_new)
    final_y_lower.append(y_range_new)
    stop_point +=1
  return final_x_lower[1:], final_y_lower[1:]
```

```
final_x_lower, final_y_lower = make_grid_circle_lower(x_range, y_range, 1000,5)
```

```
# Find upper circles
def make_grid_circle_upper(x_range, y_range, distance,num_rows):
  final_x_upper = [x_range]
  final_y_upper = [y_range]
  step = distance * math.sqrt(3) / 2
  stop_point = 0
  while stop_point <= num_rows: 
    x_range_new = [i - step for i in final_x_upper[stop_point][1:]]
    y_range_new = compute_average_cummulative(final_y_upper[stop_point])
    final_x_upper.append(x_range_new)
    final_y_upper.append(y_range_new)
    stop_point +=1
  return final_x_upper[1:], final_y_upper[1:]
```

**Define function to get venue from every neighborhood**

```
def GetNearbyVenues(names, latitudes, longitudes, radius=100, limit=100):
  venues_list = []
  for name, lat, lng in zip(names, latitudes, longitudes):
    print(name)

    # Create API request
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, limit)

    # make a GET request
    try:
      results = requests.get(url).json()["response"]["groups"][0]["items"]
    except:
      results = None

    # return only relevant information
    venues_list.append([(name, lat, lng, v["venue"]["name"], v["venue"]["location"]["lat"], v["venue"]["location"]["lng"], v["venue"]["categories"][0]["name"]) for v in results])

  # Build dataframe
  nearby_venues = pd.DataFrame(item for venue_list in venues_list for item in venue_list)

  nearby_venues.columns = ["Neighborhood", "Neighborhood Latitude", "Neighborhood Longitude", "Venue", "Venue Latitude", "Venue Longitude", "Venue Category"]

  return(nearby_venues)
```

**Define function to find out pattern about Café and Coffee Shop in cluster applied proportion in each columns**

```
def get_freqency(series, list_category):
  result = 0
  for i in list_category:
    try:
      freq = series.value_counts(normalize=True)[i]
    except:
      freq = 0
    result = result + freq
  return result

def get_choice(inp):
  result = []
  prop = [0.4, 0.3, 0.2, 0.07, 0.03]
  for num1, num2 in zip(inp, prop):
    result.append(num1 * num2)
  final = sum(result)
  return final
```

```
lst = []
list_category = ['Café', 'Coffee Shop']
for i in tes.columns:
  result = get_freqency(tes.loc[:,i], list_category) 
  lst.append(result)

result = get_choice(lst)
result
```
